<h1 style="font-family:algerian; font-size:30px; color:red">A-Normalisation Min-Max</h1>

```python
# Q10: prendre une copie de notre jeu de donnes. (df_normal).
df_normal = pd.read_csv('../data/insurance.csv')
df_normal
```

```python
# Q11: Prsentez les donnes statistiques  'df_normal'.
df_normal.describe()
```

```python
# Q12: Normalisez les deux variables quantitatives 'bmi' et 'expenses' en utilisant la formule ci-dessus.
```

```python
# 'bmi'
for i in range(df_normal.shape[0]) :
    df_normal.loc[i , 'bmi'] = ( df_normal.loc[i , 'bmi'] - df_normal['bmi'].max() ) / (df_normal['bmi'].min() - df_normal['bmi'].max() )
```

```python
# 'expenses'
for i in range(df_normal.shape[0]) :
    df_normal.loc[i , 'expenses'] = ( df_normal.loc[i , 'expenses'] - df_normal['expenses'].max() ) / (df_normal['expenses'].min() - df_normal['expenses'].max() )
```

```python
df_normal
```

```python
 # Q13: Affichez les histogrammes des deux variables, 'bmi' et 'expenses', avant et aprs la normalisation.
plt.hist(df_normal['bmi'], bins=10, color='blue', alpha=0.7)
```

<h1 style="font-family:algerian; font-size:30px; color:red">B-Standardisation</h1>

```python
# Q15: prendre une copie de notre jeu de donnes. (df_std).
data = pd.read_csv('../data/insurance.csv')
df_std = data.copy()
df_std
```

```python
# Q16: Standardisez les deux variables quantitatives 'bmi' et 'expenses' en utilisant la formule ci-dessus.
```

```python
# 'bmi'
std = np.std(df_std['bmi'])
mean = np.mean(df_std['bmi'])
for i in range(df_std.shape[0]) :
    df_std.loc[i , 'bmi'] = ( df_std.loc[i , 'bmi'] - mean ) / (std)
```

```python
# 'expenses'
std = np.std(df_std['expenses'])
mean = np.mean(df_std['expenses'])
for i in range(df_std.shape[0]) :
    df_std.loc[i , 'expenses'] = ( df_std.loc[i , 'expenses'] - mean ) / (std)

df_std
```

```python
# Q17: Affichez les histogrammes des deux variables, 'bmi' et 'expenses', avant et aprs la standardisation.
plt.figure(figsize=(12, 7))  # Taille de la figure
plt.subplot(2, 2, 1)
plt.hist(data['bmi'], bins=30, label="avant")
plt.title("bmi avant")

plt.subplot(2, 2, 2)
plt.hist(df_std['bmi'], bins=30, label="apres")
plt.title("bmi apres")

plt.subplot(2, 2, 3)
plt.hist(data['expenses'], bins=30, label="avant", color="red")
plt.title("depense avant")

plt.subplot(2, 2, 4)
plt.hist(df_std['expenses'], bins=30, label="apres", color="red")
plt.title("depense apres")
```

```python
data = pd.read_csv("../data/dataset.csv",index_col=0 ,encoding = "ISO-8859-1")
data
```

```python
data.dtypes.value_counts().plot.pie()
```

```python
data.isna().sum()
```

```python
# Suppression des colonnes avec uniquement des NaN
data = data.dropna(how='all', axis=1)  
# Suppressoin les lignes avec uniquement des NaN
data = data.dropna(how='all', axis=0) 
data
```

```python
data.shape
```

```python
# Remplacer les valeurs nulles dans chaque colonnes quantitatives par leur mediane 
data = data.apply(lambda col: col.fillna(col.median()) if col.dtype != 'object' else col)

data.head(5)
```

```python
# Remplacer les valeurs nulles dans chaque colonnes qualitatives par leur mediane 
for col in data.select_dtypes(include=['object']).columns:  
    data[col].fillna(data[col].mode()[0], inplace=True)  

data
```

```python
data.isna().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isna(), cbar=True)
plt.title("Visualisation des valeurs nulles")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    data[col] = label_encoder.fit_transform(data[col])
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
data_cleaned = pd.DataFrame(MinMaxScaler().fit_transform(data), columns=data.columns)
data_cleaned.head()
```
