```python
#importer les packages ncessaires
import pandas as pd
import numpy as np
#les packages de visualisation
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Q1 : importer les donnes
df=pd.read_csv('../data/iris.csv')
df.head()
```

```python
#Q2 : cree une variable target (flower)
flower = df['flower']
flower
```

```python
# Q4 : supprimez la colonne 'flower'.
df.drop(columns=['flower'] , inplace=True)
```

```python
# Q5 : Prsentez les donnes statistiques de notre base de donnes.
df.describe()
```

```python
# Q6 : copier le dataframe dans une nouvelle variable df_cent_reduit
df_cent_reduit = df.copy()
df_cent_reduit
```

```python
# Q7 : centraliser notre DataFrame
for col in df_cent_reduit.columns :
    
    df_cent_reduit[col] = (df_cent_reduit[col] - np.mean(df_cent_reduit[col])) / np.std(df_cent_reduit[col])

df_cent_reduit
```

```python
#importer la bibliothque PCA
from sklearn.decomposition import PCA
```

```python
# Q13 : cree un objet PCA (en pricisant n_components = NB de variabes )
pca = PCA(n_components=4)
# Q14 : entrainer le module cree (En utilisant la fonctions fit(dataset) ) / Calcul des composantes principales
pca.fit_transform(df_cent_reduit)
```

```python
# Q15 : afficher les valeurs propres (avec l'attribue explained_variance_)
pca.explained_variance_
```

```python
# Q16 : afficher les vecteurs propres (avec l'attribue components_)
pca.components_
```

```python
# Q17 : afficher la variance expliqu par chacune des composantes slectionnes ( avec l'attribue explained_variance_ratio_)
pca.explained_variance_ratio_
```

```python
# fonction pour afficher les composantes principales 
def plot_var_explique (acp):
    var_explique = acp.explained_variance_ratio_
    plt.bar(np.arange(len(var_explique ))+1, var_explique )
    plt.plot(np.arange(len(var_explique ))+1, var_explique.cumsum(),c="red",marker='o')
    plt.xlabel("la rang de l'axe d'inertie")
    plt.ylabel("pourcentage d'inertie")
    plt.title(" Eboulis des valeurs propres")
    plt.show(block=False)

```

```python
# Q18 : Eboulis des valeurs propres (En Utilisant  plot_var_explique () )
plot_var_explique(acp=pca)
```

```python
# Q19 : appliquer l'acp pour avoir 2 composants du moment que 2 composants nous permettent d'avoir 94 de l'information
# indication PCA(n_components=2).fit(data)
pca2 = PCA(n_components=2).fit(df_cent_reduit)
# Q20 : appliquer l'acp au df_cent_reduit (donnee centrer reduit) 
pca2 = pca2.transform(df_cent_reduit)
pca2
```

```python
# Q21 : dessiner les projections de NV_data sur les deux composantes qu'on a trouv en utlisant ACP
# indication (sns.scatterplot(data=dataset,x=?,y=?))
pca_df = pd.DataFrame(pca2, columns=['var 1', 'var 2'])
sns.scatterplot(data=pca_df, x='var 1', y='var 2')
```

```python
# cette fonction dessine le cercle de correlation .
def display_circles(pcs, n_comp, pca, axis_ranks, labels=None, label_rotation=0, lims=None):
    for d1, d2 in axis_ranks:
        if d2 < n_comp:

            # initialisation de la figure
            fig, ax = plt.subplots(figsize=(10,10))

            # dtermination des limites du graphique
            if lims is not None :
                xmin, xmax, ymin, ymax = lims
            elif pcs.shape[1] < 30 :
                xmin, xmax, ymin, ymax = -1, 1, -1, 1
            else :
                xmin, xmax, ymin, ymax = min(pcs[d1,:]), max(pcs[d1,:]), min(pcs[d2,:]), max(pcs[d2,:])

            # affichage des flches
            # s'il y a plus de 30 flches, on n'affiche pas le triangle  leur extrmit
            if pcs.shape[1] < 30 :
                plt.quiver(np.zeros(pcs.shape[1]), np.zeros(pcs.shape[1]),
                   pcs[d1,:], pcs[d2,:], 
                   angles='xy', scale_units='xy', scale=1, color="grey")
                # (voir la doc : https://matplotlib.org/api/_as_gen/matplotlib.pyplot.quiver.html)
            else:
                lines = [[[0,0],[x,y]] for x,y in pcs[[d1,d2]].T]
                ax.add_collection(LineCollection(lines, axes=ax, alpha=.1, color='black'))
            
            # affichage des noms des variables  
            if labels is not None:  
                for i,(x, y) in enumerate(pcs[[d1,d2]].T):
                    if x >= xmin and x <= xmax and y >= ymin and y <= ymax :
                        plt.text(x, y, labels[i], fontsize='14', ha='center', va='center', rotation=label_rotation, color="blue", alpha=0.5)
            
            # affichage du cercle
            circle = plt.Circle((0,0), 1, facecolor='none', edgecolor='b')
            plt.gca().add_artist(circle)

            # dfinition des limites du graphique
            plt.xlim(xmin, xmax)
            plt.ylim(ymin, ymax)
        
            # affichage des lignes horizontales et verticales
            plt.plot([-1, 1], [0, 0], color='grey', ls='--')
            plt.plot([0, 0], [-1, 1], color='grey', ls='--')

            # nom des axes, avec le pourcentage d'inertie expliqu
            plt.xlabel('F{} ({}%)'.format(d1+1, round(100*pca.explained_variance_ratio_[d1],1)))
            plt.ylabel('F{} ({}%)'.format(d2+1, round(100*pca.explained_variance_ratio_[d2],1)))

            plt.title("Cercle des corrlations (F{} et F{})".format(d1+1, d2+1))
            plt.show(block=False)
```
