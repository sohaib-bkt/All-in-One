```python
import matplotlib.pyplot as plt

# Donnes du portefeuille (remplacez par vos rsultats GAMS)
rendement_portefeuille = 7.8  # Rendement total du portefeuille (en %)
risque_portefeuille = 3.5     # Risque total du portefeuille (en %)

# Donnes des actions (exemple)
actions = ['CIH', 'Maroc Telecom', 'Attijariwafa Bank', 'LafargeHolcim', 'TotalEnergies']
proportions = [0.25, 0.20, 0.30, 0.15, 0.10]  # Proportions optimales

# Cration des graphiques
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Graphique 1 : Rendement et Risque du Portefeuille
ax1.bar(['Rendement', 'Risque'], [rendement_portefeuille, risque_portefeuille], color=['green', 'red'])
ax1.set_title('Rendement et Risque du Portefeuille')
ax1.set_ylabel('Valeur (%)')
ax1.set_ylim(0, 10)  # Ajustez les limites de l'axe Y si ncessaire

# Graphique 2 : Rpartition du Portefeuille par Action
ax2.pie(proportions, labels=actions, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green', 'red', 'purple'])
ax2.set_title('Rpartition du Portefeuille par Action')

# Afficher les graphiques
plt.tight_layout()
plt.show()
```
