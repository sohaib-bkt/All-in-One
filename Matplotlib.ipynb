```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

## scatter plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.scatter(x, y, color = 'red' , marker='*' , s=100 , alpha=0.5)
plt.show()
```

## line plots

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
y = [300.2 , 543.3 , 1075.9 , 2862.5 , 5979.6 , 2113.7 , 1232]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years , gdp , c='green' , lw=3 , marker='x' , ms=10 , alpha=0.5 , markeredgecolor='red' , markeredgewidth=2,label='GDP')
plt.plot(years , y ,label='sda')
plt.legend(loc='upper left')
plt.title("Nominal GDP")
plt.xlabel("Year")
plt.ylabel("Billions of $")
plt.grid(True)

plt.show()
```

## sub plots

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
y2 = [2, 5, 10, 17]

plt.subplot(2, 1, 1)  # 2 lignes, 1 colonne, 1er graphique
plt.plot(x, y, 'r')
plt.title("sin(x)")

plt.subplot(2, 1, 2)  # 2e graphique
plt.plot(x, y2, 'b')
plt.title("cos(x)")

plt.tight_layout()  # Ajuste les espacements
plt.show()
```

## bar plots

```python
x = ['c++' , 'java' , 'python' , 'c#']
y = [22 , 17 , 25 , 14]
plt.bar(x , y , color='green' , width=0.2)
plt.show()
```

## histograme

```python
ages = np.random.normal(25, 10, 1000)
plt.hist(ages)
# plt.hist(ages,cumulative=True)
plt.show()
```

## pie chart


```python
languages = ['c++' , 'java' , 'python' , 'c#']
per = [22 , 100, 25 , 14]
explode = [0.1 , 0 , 0 , 0]
plt.pie(per , labels=languages , explode=explode)
plt.show()
```

## box plot

```python
height = np.random.normal(170, 1, 300)
plt.boxplot(height)
plt.show()
```

```python


# Création d'une grille
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Fonction Z = f(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Création de la surface 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Tracé de la surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Ajout d'une barre de couleur
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.title("Surface 3D")
plt.show()
```
