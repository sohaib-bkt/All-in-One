```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
from sklearn.decomposition import PCA  
from sklearn.datasets import load_iris  

# Charger un jeu de données (par exemple, le jeu de données Iris)  
data = load_iris()  
X = data.data  
y = data.target  

# Normaliser les données (facultatif mais recommandé)  
from sklearn.preprocessing import StandardScaler  
X_scaled = StandardScaler().fit_transform(X)  

# Appliquer PCA  
pca = PCA(n_components=2)  # Nous choisissons de réduire à 2 dimensions  
X_pca = pca.fit_transform(X_scaled)  

# Créer un DataFrame pour faciliter la visualisation  
df_pca = pd.DataFrame(data=X_pca, columns=['Composant 1', 'Composant 2'])  
df_pca['Cible'] = y  

# Visualiser les résultats  
plt.figure(figsize=(8, 6))  
colors = ['red', 'green', 'blue']  
for target, color in zip(np.unique(y), colors):  
    plt.scatter(df_pca[df_pca['Cible'] == target]['Composant 1'],  
                df_pca[df_pca['Cible'] == target]['Composant 2'],  
                color=color, label=data.target_names[target])  

plt.title('Analyse en Composants Principaux (PCA)')  
plt.xlabel('Composant 1')  
plt.ylabel('Composant 2')  
plt.legend()  
plt.grid()  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du signal
A = 5  # Amplitude
omega = 2 * np.pi * 50  # Pulsation (fréquence de 50 Hz)
phi = np.pi / 4  # Phase initiale
t = np.linspace(0, 0.02, 1000)  # Temps (20 ms)

# Calcul des composantes
x_real = A * np.cos(omega * t + phi)  # Partie réelle
x_imag = A * np.sin(omega * t + phi)  # Partie imaginaire

# Signal complexe
x_complex = x_real + 1j * x_imag

# Tracé des composantes
plt.figure(figsize=(10, 6))
plt.plot(t, x_real, label="Partie Réelle (Re(x(t)))")
plt.plot(t, x_imag, label="Partie Imaginaire (Im(x(t)))")
plt.title("Signal complexe et ses composantes")
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
