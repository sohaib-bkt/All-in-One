```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres du signal
V0 = 1.0  # Composante continue
Vn = [0.5, 0.3, 0.2]  # Amplitudes des harmoniques
theta_n = [0, np.pi/4, np.pi/2]  # Phases des harmoniques
Fp = 5  # Frquence fondamentale en Hz
t = np.linspace(0, 1, 500)  # Vecteur de temps de 0  1 seconde

# Fonction de reconstruction du signal
def reconstruct_signal(t, V0, Vn, theta_n, Fp):
    signal = V0 * np.ones_like(t)  # Initialiser avec la composante continue
    for n, (V, theta) in enumerate(zip(Vn, theta_n), start=1):
        signal += V * np.cos(2 * np.pi * n * Fp * t + theta)
    return signal

# Reconstruction du signal
x_t = reconstruct_signal(t, V0, Vn, theta_n, Fp)

# Visualisation du signal
plt.figure(figsize=(10, 4))
plt.plot(t, x_t, label='Signal reconstruit')
plt.title('Reconstruction d\'un signal temporel  partir de ses composantes frquentielles')
plt.xlabel('Temps [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

## serie de fourier

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Nombre de coefficients  
N = 20  

# Calcul des coefficients de Fourier  
frequencies = np.arange(1, N+1)  
a_n = (1 - (-1)**frequencies) / (4 * np.pi**2 * frequencies**2)  # Coefficients pour n impair  
amplitudes = np.zeros(N)  

# On remplit les amplitudes pour les frquences paires et impaires  
for n in frequencies:  
    if n % 2 == 1:  # uniquement les n impairs  
        amplitudes[n - 1] = a_n[n - 1]  

# Reprsentation en domaine frquentiel  
plt.figure(figsize=(10, 6))  
plt.stem(frequencies, amplitudes, basefmt=" ")  
plt.title('Spectre de Fourier de la fonction f(t)')  
plt.xlabel('Frquence (n)')  
plt.ylabel('Amplitude')  
plt.xticks(frequencies)  # pour afficher toutes les frquences  
plt.grid()  
plt.show()
```

## transformer de fourier 


```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres du signal
f0 = 5  # Frquence du cosinus (Hz)
T = 1.0 / f0  # Priode du cosinus
fs = 1000  # Frquence d'chantillonnage (Hz)
t = np.linspace(0, 1, int(fs), endpoint=False)  # Temps de 0  1 seconde

# Signal cosinus
x = np.cos(2 * np.pi * f0 * t)

# Calcul de la transforme de Fourier
X = np.fft.fft(x)  # Transforme de Fourier discrte
frequencies = np.fft.fftfreq(len(x), d=1/fs)  # Axe des frquences

# Affichage du signal temporel
plt.figure(figsize=(12, 6))

# Signal temporel
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title(f"Signal cosinus : $x(t) = \cos(2\pi f_0 t)$ avec $f_0 = {f0}$ Hz")
plt.xlabel("Temps [s]")
plt.ylabel("Amplitude")
plt.grid()

# Transforme de Fourier
plt.subplot(2, 1, 2)
plt.stem(frequencies, np.abs(X) / len(x), markerfmt=" ", basefmt=" ")
plt.title("Transforme de Fourier du cosinus")
plt.xlabel("Frquence [Hz]")
plt.ylabel("Amplitude")
plt.xlim(-2 * f0, 2 * f0)  # Zoom autour des frquences f0
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres du signal
T = 2  # Largeur de la porte (en secondes)
fs = 1000  # Frquence d'chantillonnage (Hz)
t = np.linspace(-5, 5, int(fs), endpoint=False)  # Temps de -5  5 secondes

# Signal rectangulaire
x = np.where(np.abs(t) <= T / 2, 1, 0)

# Calcul de la transforme de Fourier
X = np.fft.fft(x)  # Transforme de Fourier discrte
frequencies = np.fft.fftfreq(len(x), d=1/fs)  # Axe des frquences

# Normalisation de la transforme de Fourier
X = np.fft.fftshift(X)  # Centrer le spectre autour de 0 Hz
frequencies = np.fft.fftshift(frequencies)  # Centrer les frquences

# Affichage du signal temporel
plt.figure(figsize=(12, 6))

# Signal temporel
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title(f"Signal rectangulaire de largeur $T = {T}$ s")
plt.xlabel("Temps [s]")
plt.ylabel("Amplitude")
plt.grid()

# Transforme de Fourier
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.abs(X) / len(x))
plt.title("Transforme de Fourier du signal rectangulaire")
plt.xlabel("Frquence [Hz]")
plt.ylabel("Amplitude")
plt.xlim(-5, 5)  # Zoom autour de 0 Hz
plt.grid()

plt.tight_layout()
plt.show()
```

## convulation

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres
T = 2  # Largeur de la fonction rectangulaire
t = np.linspace(-5, 5, 1000)  # Axe des temps

# Fonctions
x_t = np.ones_like(t)  # Fonction constante x(t) = 1
y_t = np.where(np.abs(t) <= T / 2, 1, 0)  # Fonction rectangulaire y(t)

# Convolution
z_t = T * np.ones_like(t)  # Rsultat de la convolution z(t) = T

# Affichage
plt.figure(figsize=(10, 6))

# Fonction constante x(t)
plt.subplot(3, 1, 1)
plt.plot(t, x_t, label="x(t) = 1")
plt.title("Fonction constante x(t)")
plt.xlabel("Temps [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# Fonction rectangulaire y(t)
plt.subplot(3, 1, 2)
plt.plot(t, y_t, label="y(t) = rect(t/T)")
plt.title("Fonction rectangulaire y(t)")
plt.xlabel("Temps [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# Convolution z(t)
plt.subplot(3, 1, 3)
plt.plot(t, z_t, label="z(t) = x(t) * y(t)")
plt.title("Convolution z(t)")
plt.xlabel("Temps [s]")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

## echant


```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres
f0 = 10  # Frquence du signal (Hz)
fs = 25  # Frquence d'chantillonnage (Hz)
Ts = 1 / fs  # Priode d'chantillonnage (s)
n = np.arange(0, 50)  # Indices d'chantillonnage

# Signal chantillonn
x_n = np.cos(2 * np.pi * f0 * n * Ts)

# Affichage
plt.figure(figsize=(10, 4))
plt.stem(n, x_n, basefmt=" ")
plt.title(f"Signal chantillonn  $f_s = {fs}$ Hz")
plt.xlabel("Indice d'chantillonnage $n$")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```

```python
# Paramtres
fs_aliasing = 15  # Frquence d'chantillonnage avec aliasing (Hz)
Ts_aliasing = 1 / fs_aliasing  # Priode d'chantillonnage (s)

# Signal chantillonn avec aliasing
x_n_aliasing = np.cos(2 * np.pi * f0 * n * Ts_aliasing)

# Affichage
plt.figure(figsize=(10, 4))
plt.stem(n, x_n_aliasing, basefmt=" ")
plt.title(f"Signal chantillonn  $f_s = {fs_aliasing}$ Hz (aliasing)")
plt.xlabel("Indice d'chantillonnage $n$")
plt.ylabel("Amplitude")
plt.grid()
plt.show()
```
