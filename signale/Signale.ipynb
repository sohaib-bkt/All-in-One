```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres temporels
t = np.linspace(0, 0.0003, 1000)  # 0.3ms pour voir plusieurs priodes

# Cration des signaux
s1 = np.full_like(t, 4)  # Signal constant  4V
s2 = 4 + 3 * np.cos(2 * np.pi * 10**4 * t)  # Signal sinusodal

# Cration de la figure avec deux sous-graphiques
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Premier signal
ax1.plot(t*1000, s1, 'b-', label='s1(t)')
ax1.set_title('Signal s1(t)')
ax1.set_xlabel('Temps (ms)')
ax1.set_ylabel('Tension (V)')
ax1.grid(True)
ax1.legend()
ax1.set_ylim(0, 8)

# Deuxime signal
ax2.plot(t*1000, s2, 'r-', label='s2(t)')
ax2.set_title('Signal s2(t)')
ax2.set_xlabel('Temps (ms)')
ax2.set_ylabel('Tension (V)')
ax2.grid(True)
ax2.legend()
ax2.set_ylim(0, 8)

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```
